```python
import urllib.request, urllib.error, urllib.parse
import json
import os
from pprint import pprint
```

```python

REST_URL = "http://data.bioontology.org"
API_KEY = "70d0c0ad-f6be-48e2-9241-8d987f4d1037"

def get_json(url):
    opener = urllib.request.build_opener()
    opener.addheaders = [('Authorization', 'apikey token=' + API_KEY)]
    return json.loads(opener.open(url).read())
```

```python

# Get list of search terms
path = os.path.join(os.path.dirname('/home/mamonteiro/Project-LEI/OFFSIDES/Bioportal/'), 'classes_search_terms.txt')
terms_file = open(path, "r")
terms = []
for line in terms_file:
    terms.append(line)
```

```python

# Do a search for every term
search_results = []
for term in terms:
    search_results.append(get_json(REST_URL + "/search?q=" + term)["collection"])

# Print the results
for result in search_results:
    pprint(result)
```
